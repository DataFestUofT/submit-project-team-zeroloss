```python
from bs4 import BeautifulSoup
import spacy
import unidecode
from word2number import w2n
from pycontractions import Contractions
import gensim.downloader as api
import pandas as pd
```

```python
nlp = spacy.load('en_core_web_md')
print("loaded en_core_web_md")
```

```python
model = api.load("glove-twitter-25")
print("loaded api model")
```

```python
# exclude words from spacy stopwords list
deselect_stop_words = ['no', 'not']
for w in deselect_stop_words:
    nlp.vocab[w].is_stop = False
cont = Contractions(kv_model=model)
cont.load_models()
```

```python
data1 = pd.read_csv("../semeval2017/twitter-2013train-A.tsv", sep="\t", header=None)
```

```python
data2 = pd.read_csv("../semeval2017/twitter-2015train-A.tsv", sep="\t", header=None)
```

```python
data3 = pd.read_csv("../semeval2017/twitter-2016train-A.tsv", sep="\t", header=None)
```

```python
data1['text'] = list(cont.expand_texts(list(data1[2]), precise=True))
```

```python
data2['text'] = list(cont.expand_texts(list(data2[2]), precise=True))
```

```python
data3['text'] = list(cont.expand_texts(list(data3[2]), precise=True))
```

```python
for s in data1['text']:
    print(s)
```

```python
data1.to_csv("data1.csv")
data2.to_csv("data2.csv")
data3.to_csv("data3.csv")
```
