```python
import pandas as pd

df = pd.read_csv("../data/hydrated_data_ieee/final_data/shallow_processed_with_prediction.csv")
```

```python
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
```

```python
import demoji
demoji.download_codes()
```

```python
def check_contains_emoji(text):
    if not isinstance(text, str): # handle NaN
        return False
    else:
        def deEmojify(inputString):
#             print(inputString, type(inputString))
#             raise Exception()
            return inputString.encode('ascii', 'ignore').decode('ascii')
        return deEmojify(text) != text
```

```python
df["username_contains_emoji"] = df["user.name"].map(check_contains_emoji)
```

```python
df["userscreenname_contains_emoji"] = df["user.screen_name"].map(check_contains_emoji)
```

```python
df["userdescription_contains_emoji"] = df["user.description"].map(check_contains_emoji)
```
